```python
import pandas as pd

# Read the CSV file into a DataFrame
da = pd.read_csv('electiondata4.csv')

# Filter the DataFrame to include only rows where the 'won status' column is 'won'
won_data = da[da['won status'] == 'won']

# Sort the filtered DataFrame by the 'State Name' column
won_data_sorted = won_data.sort_values(by='state')

# Function to filter data based on state name
def filter_by_state(state_name):
    state_wins = won_data_sorted[won_data_sorted['state'] == state_name]
    return state_wins

# Get user input for the state name
state_name_input = input("Enter the state name: ")

# Filter the data based on the user input
state_wins_data = filter_by_state(state_name_input)

# Display the filtered data
print(state_wins_data)
```

```python
import pandas as pd
da = pd.read_csv('electiondata4.csv')
won_data = da[da['won status'] == 'won'] 
won_data_sorted = won_data.sort_values(by='state')
def get_winners_by_state(state_name):
    state_winners = won_data_sorted[won_data_sorted['state'] == state_name]
    return state_winners[['won status', 'Party Name']]
state_name_input = input("Enter the state name: ")
state_winners_data = get_winners_by_state(state_name_input)
party_counts = state_winners_data['Party Name'].value_counts()
for party, count in party_counts.items():
    print(f'Party: {party}, Total Wins: {count}')
```
